## LSTM for Differenced Data

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# Load and preprocess data
dataset_train = pd.read_csv('/kaggle/input/differenced-labelled/difference_data (1).csv')
dataset_test = pd.read_csv('/kaggle/input/differenced-labelled/difference_data_test (1).csv')

scaler = MinMaxScaler(feature_range=(0, 1))

# Train data
X_train = scaler.fit_transform(dataset_train[['open', 'close', 'high', 'low']].values)
y_train = dataset_train['pattern'].values

# Test data
X_test = scaler.transform(dataset_test[['open', 'close', 'high', 'low']].values)
y_test = dataset_test['pattern'].values

# Reshape the input data for LSTM
X_train = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_test = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))

# Define LSTM model
model = Sequential([
    LSTM(64, input_shape=(X_train.shape[1], X_train.shape[2]), return_sequences=True),
    Dropout(0.2),
    LSTM(64, return_sequences=True),
    Dropout(0.2),
    LSTM(64),
    Dense(5, activation='softmax')
])

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=10, batch_size=64)
```

## Display of Evaluation Parameters

```python
# Predict probabilities for each class
probabilities = model.predict(X_test)

# Convert probabilities to class labels
y_pred = np.argmax(probabilities, axis=1)

# Print evaluation report
report = classification_report(y_test, y_pred, output_dict=True)

# Extract required metrics
accuracy = report['accuracy']
precision = report['macro avg']['precision']
recall = report['macro avg']['recall']
f1_score = report['macro avg']['f1-score']

# Print the metrics
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1_score)
```
