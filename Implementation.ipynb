```python
import pandas as pd
import os
import matplotlib.pyplot as plt

# Load sample data
sample_data = pd.read_csv('SampleData.csv')

sample_data['date'] = pd.to_datetime(sample_data['date'])

# Load ground truth data
ground_truth = pd.read_csv('SampleDataGT.csv')

# Create output directory if it doesn't exist
output_dir = 'output_plots'
os.makedirs(output_dir, exist_ok=True)

# Iterate through each range and pattern
for index, row in ground_truth.iterrows():
    start_time = pd.to_datetime(row['start'])
    end_time = pd.to_datetime(row['end'])
    pattern = row['pattern']
    
    # Filter sample data based on the range
    filtered_data = sample_data[(sample_data['date'] >= start_time) & (sample_data['date'] <= end_time)]
    
    # Plot the graph
    plt.figure(figsize=(10, 6))
    plt.plot(filtered_data['date'], filtered_data['open'])
    plt.title(f'{pattern} - Range: {start_time} to {end_time}')
    plt.xlabel('Date')
    plt.ylabel('Open Value')
    plt.grid(True)
    
    # Create folder for each pattern if it doesn't exist
    pattern_dir = os.path.join(output_dir, pattern)
    os.makedirs(pattern_dir, exist_ok=True)
    
    # Replace special characters in the file name
    start_time_str = str(start_time).replace(':', '_')
    end_time_str = str(end_time).replace(':', '_')
    plot_filename = f'{start_time_str}_{end_time_str}.png'
    
    # Save the plot in the corresponding folder
    plot_filepath = os.path.join(pattern_dir, plot_filename)
    plt.savefig(plot_filepath)
    plt.close()
    
    print(f'Plot saved: {plot_filepath}')
```

```python
import os
import shutil

# Define directories
dataset_dir = 'output_plots'  # Root directory containing class directories
train_dir = 'train_data'  # Directory for training data
validation_dir = 'validation_data'  # Directory for validation data

# Create train and validation directories if they don't exist
os.makedirs(train_dir, exist_ok=True)
os.makedirs(validation_dir, exist_ok=True)

# Function to count images in each class directory
def count_images(class_dir):
    count = sum([len(files) for _, _, files in os.walk(class_dir)])
    return count

# Iterate through each class directory
for class_name in os.listdir(dataset_dir):
    class_dir = os.path.join(dataset_dir, class_name)
    
    # Count the number of images in the class directory
    num_images = count_images(class_dir)
    print(f"Class '{class_name}' has {num_images} images.")
    
    # Calculate split ratio (80% train, 20% validation)
    num_train_images = int(0.8 * num_images)
    num_validation_images = num_images - num_train_images
    
    # Create train and validation directories for the class
    train_class_dir = os.path.join(train_dir, class_name)
    os.makedirs(train_class_dir, exist_ok=True)
    validation_class_dir = os.path.join(validation_dir, class_name)
    os.makedirs(validation_class_dir, exist_ok=True)
    
    # Move images to train and validation directories
    images = os.listdir(class_dir)
    train_images = images[:num_train_images]
    validation_images = images[num_train_images:]
    
    for image in train_images:
        src = os.path.join(class_dir, image)
        dst = os.path.join(train_class_dir, image)
        shutil.move(src, dst)
    
    for image in validation_images:
        src = os.path.join(class_dir, image)
        dst = os.path.join(validation_class_dir, image)
        shutil.move(src, dst)

print("Data split into train and validation sets successfully.")
```

```python
import pandas as pd
import os
import matplotlib.pyplot as plt

# Load sample data
sample_data = pd.read_csv('SampleData.csv')
sample_data['date'] = pd.to_datetime(sample_data['date'])

# Load ground truth data
ground_truth = pd.read_csv('SampleDataGT.csv')
ground_truth['start'] = pd.to_datetime(ground_truth['start'])
ground_truth['end'] = pd.to_datetime(ground_truth['end'])

# Create output directory if it doesn't exist
output_dir = 'output_plots'
os.makedirs(output_dir, exist_ok=True)

# Iterate through each range in ground truth data
for i in range(len(ground_truth) - 1):
    end_time = ground_truth.loc[i, 'end']
    start_time_next = ground_truth.loc[i + 1, 'start']

    # Check if there's a gap between the current end time and the next start time
    if end_time < start_time_next:
        # Filter sample data based on the range
        filtered_data = sample_data[(sample_data['date'] >= end_time) & (sample_data['date'] <= start_time_next)]
        
        # Plot the graph
        plt.figure(figsize=(10, 6))
        plt.plot(filtered_data['date'], filtered_data['open'])
        plt.title(f'Others - Range: {end_time} to {start_time_next}')
        plt.xlabel('Date')
        plt.ylabel('Open Value')
        plt.grid(True)
        
        # Create 'others' folder if it doesn't exist
        others_dir = os.path.join(output_dir, 'others')
        os.makedirs(others_dir, exist_ok=True)
        
        # Save the plot in the 'others' folder
        plot_filename = f'{end_time.strftime("%Y-%m-%d_%H-%M-%S")}_to_{start_time_next.strftime("%Y-%m-%d_%H-%M-%S")}.png'
        plot_filepath = os.path.join(others_dir, plot_filename)
        plt.savefig(plot_filepath)
        plt.close()
        
        print(f'Plot saved: {plot_filepath}')
```

```python
import os
import shutil
import random

# Define paths
output_folder = 'output_plots/others'
train_data_folder = 'train_data/others'
validation_data_folder = 'validation_data/others'

# Create directories for training and validation data if they don't exist
os.makedirs(train_data_folder, exist_ok=True)
os.makedirs(validation_data_folder, exist_ok=True)

# Get a list of image files in the output folder
image_files = os.listdir(output_folder)

# Define the ratio of images to be used for training (e.g., 80%)
train_ratio = 0.8

# Calculate the number of images for training and validation
num_images = len(image_files)
num_train_images = int(train_ratio * num_images)
num_validation_images = num_images - num_train_images

# Shuffle the list of image files
random.shuffle(image_files)

# Move images to training directory
for i in range(num_train_images):
    src = os.path.join(output_folder, image_files[i])
    dst = os.path.join(train_data_folder, image_files[i])
    shutil.move(src, dst)
    print(f'Moved {image_files[i]} to training data.')

# Move remaining images to validation directory
for i in range(num_train_images, num_images):
    src = os.path.join(output_folder, image_files[i])
    dst = os.path.join(validation_data_folder, image_files[i])
    shutil.move(src, dst)
    print(f'Moved {image_files[i]} to validation data.')
```
